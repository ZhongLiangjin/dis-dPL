# Plot ensemble mean performance of three tramsfer learning models (tDPL-a, tDPL-b, tDPL-c) for comparison.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from collections import defaultdict
from ensembleUtils import Evaluator
import seaborn as sns
import pandas as pd
from pathlib import Path
```

## Get metrics for streamflow, ET and snow depth for each model during 2000-2019

```python
# load four transfer learning models
rootDir = '../checkpoints/Types'
meanOpt = True  # determine whether plot mean performance for ET and snow depth
modelNames = ['TL-a', 'TL-b', 'TL-c', 'TL-d']
stations = ['TNH', 'JUG', 'MAQ', 'MET', 'JIM']
#  initiate a dict to store NSE, r, and PBIAS of streamflow, r and RMSE of ET and snow depth for five models
modelMetric, simulation = defaultdict(dict), dict()
for model in modelNames:
    print(f'Calculating metrics for model {model}')
    modelDir = os.path.join(rootDir, model)
    evaluator = Evaluator(ensembleFolder=modelDir,
                          basinFile='../data/sub-basins/watershed.shp',
                          flowFile='../data/streamflow.csv',
                          valDataFile='../data/valData.pkl',
                          modelTyp='dDPL')
    modelMetric[model] = evaluator.metric
    simulation[model] = evaluator.sim

seeds = [k for k in modelMetric[modelNames[0]]['Q'].keys() if k is not 'mean']
```

### Plot box diagram for streamflow

```python
plt.rcParams['font.family'] = 'Arial'
sns.set(style="ticks")

def auto_text(rects, ax, metric):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            xy=(rect.get_x() + rect.get_width() / 2, height)
        else:
            xy=(rect.get_x() + rect.get_width() / 2, 0)

        text = f'{height: .1f}' if metric in ['PBIAS'] else f'{height: .3f}'
        ax.annotate(text,  # put the detail data
                    xy=xy,  # get the center location.
                    xytext=(0, 1),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

fig = plt.figure(figsize=(12, 5), dpi=300)

# 'T' indicates temporal test station, 'S' indicates spatial test stations
markers = {
    'TL-a': {'TNH': 'T', 'MAQ': 'T', 'JIM': 'T', 'MET': 'S', 'JUG': 'S'},
    'TL-b': {'TNH': 'T', 'MAQ': 'T', 'JIM': 'T', 'MET': 'S', 'JUG': 'S'},
    'TL-c': {'TNH': 'T', 'MAQ': 'T', 'JIM': 'T', 'MET': 'S', 'JUG': 'S'},
    'TL-d': {'TNH': 'T', 'MAQ': 'T', 'JIM': 'T', 'MET': 'S', 'JUG': 'S'}
}
rename_dict = dict(zip(['TL-a', 'TL-b', 'TL-c', 'TL-d'], ['dis-TL-3a', 'dis-TL-3b', 'dis-TL-3c', 'dis-TL-3d']))
metricUnit = dict(zip(['NSE', 'RMSE'], ['', '(mm/d)']))
colors = dict(zip(markers.keys(), [c for c in sns.color_palette('pastel', n_colors=4)]))

for i, metricName in enumerate(metricUnit.keys()):
    yMax, yMin = -9999, 9999
    ax = fig.add_subplot(2, 1, i + 1)
    boxes = defaultdict()
    space, width = 0.1, 0.2
    for k, station in enumerate(stations):
        for j, model in enumerate(markers.keys()):
            xTemp = np.array([(space + width) * (0.5 + j + k * 4) + space * k * 1.5])
            yTemp = np.array([modelMetric[model]['Q']['mean'][station][metricName]])
            yMax = yTemp.max() if yTemp.max() > yMax else yMax
            yMin = yTemp.min() if yTemp.min() < yMin else yMin
            rect = ax.bar(xTemp, yTemp, width, color=colors[model], label=rename_dict[model], alpha=0.8)
            auto_text(rect, ax, metricName)

    # ax.set_xlabel(station, labelpad=2.5)
    top = min(yMax * 1.6, 1.0) if metricName in ['NSE', 'r', 'KGE'] else yMax * 1.3
    bottom = yMin * 0.8 if yMin > 0 else yMin * 1.3
    ax.set_ylim(top=top, bottom=bottom)
    xticks = [(space + width) * (2 + k * 4) + space * k * 1.5 for k in range(len(stations))]
    if i == 1:
        ax.set_xticks(xticks)
        ax.set_xticklabels(stations)
    else:
        ax.set_xticks([])

    # add scatter plot
    for k, station in enumerate(stations):
        for j, model in enumerate(markers.keys()):
            xScatter = (space + width) * (0.5 + j + k * 4) + space * k * 1.5
            yScatter = modelMetric[model]['Q']['mean'][station][metricName]
            yScatter = (yScatter - max(bottom, 0)) * 0.5 + max(bottom, 0)
            ax.annotate(markers[model][station], xy=(xScatter, yScatter), xytext=(0, -1.5), ha='center', va='center', fontsize=9, color='black', textcoords='offset points', alpha=0.8)
            if j == 3:
                ax.axvline(x=xScatter + 1.25 * space + width / 2, color='grey', linestyle='--', alpha=0.5, linewidth=1)

    ax.tick_params(axis='y', length=3)  # adjust tick length
    ax.tick_params(axis='x', length=3, color='white')  # adjust tick length
    ax.set_ylabel(metricName + metricUnit[metricName])
    ax.set_xlim(left=xticks[0] - 2 * (width + space) - 0.75 * space, right=xticks[-1] + 2 * (width + space) + 0.75 * space)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    # ax.grid(axis='y',ls='--',alpha=0.8)
    if i == 0:
        ax.annotate(f'{chr(97 + i)}. {metricName} on test set ', xy=(0.87, 0.95), xycoords='axes fraction')
    else:
        ax.annotate(f'{chr(97 + i)}. {metricName} on test set ', xy=(0.858, 0.95), xycoords='axes fraction')

lines, labels = ax.get_legend_handles_labels()
fig.legend(lines[:4], labels[:4], ncol=4, borderaxespad=0.1,
           bbox_to_anchor=(0.3, 0.03, 0.4, 0.02), mode='expand', columnspacing=1, frameon=False, handletextpad=0.4)

fig.tight_layout(rect=(0, 0.04, 1, 1))
# plt.savefig(f'{rootDir}/runoff_TL.pdf')
```

### compare four transfer learning models' surface flow and subsurface flow with distributed dPL model

```python
evaluator = Evaluator(ensembleFolder='../checkpoints/dis-dPL-3',
                      basinFile='../data/sub-basins/watershed.shp',
                      flowFile='../data/streamflow.csv',
                      valDataFile='../data/valData.pkl',
                      modelTyp='dDPL')
modelMetric['dis-dPL-3'] = evaluator.metric
simulation['dis-dPL-3'] = evaluator.sim
```

```python
import geopandas as gpd
from matplotlib import colors as cm
def plotMap(simulation, cmaps, basinFile, figSize=(10, 8.5), savePath=None):
    plt.rcParams['font.family'] = 'Arial'
    sns.set(style='ticks')
    basins = gpd.read_file(basinFile)
    # metricUnit = dict(zip(['Qs', 'Qb'], ['surface flow (mm/d)', 'subsurface flow (mm/d)']))
    metricUnit = dict(zip(['Qs', 'Qb'], ['$Q_{surf}\ (mm/d)$', '$Q_{sub}\ (mm/d)$']))
    fig, axes = plt.subplots(nrows=4, ncols=3, figsize=figSize, dpi=300)
    for i, var in enumerate(metricUnit.keys()):
        # calculate data to plot
        rename_dict = dict(zip(['TL-a', 'TL-b', 'TL-c', 'TL-d', 'dis-dPL-3'], ['dis-TL-3a', 'dis-TL-3b', 'dis-TL-3c', 'dis-TL-3d', 'dis-dPL-3']))
        df = pd.DataFrame(columns=rename_dict.values(), index=range(0, 69))
        for model in rename_dict.keys():
            if var == 'Qs':
                # df.loc[:, rename_dict[model]] = simulation[model]['mean'][var].groupby(simulation[model]['mean'][var].index.year).mean().mean(axis=0)
                dfTemp = simulation[model]['mean'][var][(simulation[model]['mean'][var].index.month >= 5) & (simulation[model]['mean'][var].index.month <= 10)]
                df.loc[:, rename_dict[model]] = dfTemp.groupby(dfTemp.index.year).mean().mean(axis=0)
            else:
                dfTemp = simulation[model]['mean'][var][(simulation[model]['mean'][var].index.month >= 11) | (simulation[model]['mean'][var].index.month <= 4)]
                df.loc[:, rename_dict[model]] = dfTemp.groupby(dfTemp.index.year).mean().mean(axis=0)

        for j, item in enumerate(rename_dict.keys()):
            cmap = sns.color_palette(cmaps[i], as_cmap=True)
            ax = axes[i*2+j//3, j%3]
            dfPlot = pd.concat([basins, df], axis=1)
            vmin, vmax = df.loc[:, rename_dict[item]].values.min(), df.loc[:, rename_dict[item]].values.max()
            dfPlot.plot(column=rename_dict[item], ax=ax, cmap=cmap, edgecolor='None', linewidth=0.1, vmin=vmin, vmax=vmax)
            ax.set_xticks([])
            ax.set_yticks([])

            # Set axis spines properties
            lw = 0.8
            ax.spines['top'].set_linewidth(lw)  # Top spine thickness
            ax.spines['bottom'].set_linewidth(lw)  # Bottom spine thickness
            ax.spines['left'].set_linewidth(lw)  # Left spine thickness
            ax.spines['right'].set_linewidth(lw)  # Right spine thickness

            text = rf'{rename_dict[item]} {metricUnit[var]}'
            ax.annotate(f'{chr(97+i*4+j)}. {text}', xy=(0.01, 1.04), xycoords='axes fraction', fontsize=11)
            norm = cm.Normalize(vmin=vmin, vmax=vmax)
            sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
            cbar = plt.colorbar(sm, ax=ax, orientation='vertical', shrink=0.88, pad=0.02)
            cbar.ax.tick_params(direction='out', length=2, labelsize=10)
            cbar.outline.set_linewidth(lw)

    plt.subplots_adjust(wspace=0.05, hspace=0.1)
    plt.delaxes(ax=axes[1, 2])
    plt.delaxes(ax=axes[3, 2])
    # plt.tight_layout()
    if savePath is not None:
        plt.savefig(savePath)
    plt.show()

plotMap(simulation=simulation, cmaps=['light:b', 'light:b'], basinFile='../data/sub-basins/watershed.shp')
```

```python
import geopandas as gpd
from matplotlib import colors as cm

def plotMap(simulation, cmaps, basinFile, figSize=(10, 4.2), savePath=None):
    plt.rcParams['font.family'] = 'Arial'
    sns.set(style='ticks')
    basins = gpd.read_file(basinFile)
    # metricUnit = dict(zip(['Qs', 'Qb'], ['surface flow (mm/d)', 'subsurface flow (mm/d)']))
    metricUnit = dict(zip(['Qs', 'Qb'], ['$Q_{surf} (mm/d)$', '$Q_{sub} (mm/d)$']))
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=figSize, dpi=300)
    # calculate data to plot
    rename_dict = dict(zip(['TL-a', 'TL-b', 'TL-c', 'TL-d', 'dis-dPL-3'], ['dis-TL-3a', 'dis-TL-3b', 'dis-TL-3c', 'dis-TL-3d', 'dis-dPL-3']))
    df = pd.DataFrame(columns=rename_dict.values(), index=range(0, 69))
    for model in rename_dict.keys():
        dfTemp = simulation[model]['mean']['Qs'] + simulation[model]['mean']['Qb']
        df.loc[:, rename_dict[model]] = dfTemp.groupby(dfTemp.index.year).mean().mean(axis=0)

    for j, item in enumerate(rename_dict.keys()):
        # cmap = sns.color_palette(cmaps, as_cmap=True)
        cmap = sns.color_palette("ch:start=.2,rot=-.3", as_cmap=True)
        ax = axes[j//3, j%3]
        dfPlot = pd.concat([basins, df], axis=1)
        vmin, vmax = df.loc[:, rename_dict[item]].values.min(), df.loc[:, rename_dict[item]].values.max()
        dfPlot.plot(column=rename_dict[item], ax=ax, cmap=cmap, edgecolor='None', linewidth=0.1, vmin=vmin, vmax=vmax)
        ax.set_xticks([])
        ax.set_yticks([])

        # Set axis spines properties
        lw = 0.8
        ax.spines['top'].set_linewidth(lw)  # Top spine thickness
        ax.spines['bottom'].set_linewidth(lw)  # Bottom spine thickness
        ax.spines['left'].set_linewidth(lw)  # Left spine thickness
        ax.spines['right'].set_linewidth(lw)  # Right spine thickness

        text = rf'{rename_dict[item]}'
        ax.annotate(f'{chr(97+j)}. {text}', xy=(0.01, 1.04), xycoords='axes fraction', fontsize=11)
        # norm = cm.Normalize(vmin=vmin, vmax=vmax)
        norm = cm.Normalize(vmin=0, vmax=1)
        sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
        cbar = plt.colorbar(sm, ax=ax, orientation='vertical', shrink=0.86, pad=0.02)
        cbar.ax.tick_params(direction='out', length=2, labelsize=10)
        cbar.outline.set_linewidth(lw)

    plt.subplots_adjust(wspace=0.05, hspace=0.1)
    plt.delaxes(ax=axes[1, 2])
    # plt.tight_layout()
    if savePath is not None:
        plt.savefig(savePath)
    plt.show()

plotMap(simulation=simulation, cmaps='light:b', basinFile='../data/sub-basins/watershed.shp', savePath='E:/Research/dPL/thesis/figures/Revision1/Fig7/figure.pdf')
```
